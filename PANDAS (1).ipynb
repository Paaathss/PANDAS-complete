pipe Method



```python
import pandas as pd
import numpy as np

data01 = [[40,50,70,20,30,10],[20,30,40,50,60,70],[50,60,70,80,90,30]]

df = pd.DataFrame(data01)

def adder(a,b):
  return a+b

print(df)

print('\n')

df1 = df.pipe(adder,2)

print(df1)
```

```python
import pandas as pd
import numpy as np
data = [[40,60,30,40,50,12,451],[40,60,88,98,12,47],[40,50,60,70,90,88]]
df = pd.DataFrame(data)
def adder(a,b):
  return a+b
print(df)
df1 = df.pipe(adder,3)
print(df1)

```

AXIS AND APPLY FUNCTION



```python
import pandas as pd
import numpy as np
data = [[40,60,30,40,50,12,451],[40,60,88,98,12,47],[40,50,60,70,90,88]]
df = pd.DataFrame(data)
def adder(a):
  return a**2
print(df)
df1 = df.apply(adder,axis=0)    # column wise
print(df1)

```

```python
import pandas as pd
import numpy as np
data = [[40,60,30,40,50,12,451],[40,60,88,98,12,47],[40,50,60,70,90,88]]
df = pd.DataFrame(data)
def adder(a):
  return a**2
print(df)
df1 = df.apply(adder,axis=1)   #row wise
print(df1)

```
